## 2.1 基本数据类型
变量可以是数学里的变量，也可以是任何存储数据的单元，变量名称是大小写敏感的。

```python
count = 100     #整数型
miles = 100.0   #float
name = 'Jack'   #字符串,使用英文引号
right = True    # bool类型，对或错，这不是字符串
ca = count/8
cb = miles/8
print(ca)
print(cb)
print(name)
print(right)
```

- ‘#’开启注释，至行尾结束
- 注释内容不会被执行，仅用于解释说明，提高代码的可读性
- 在运算中整数被自动升级为小数
- 尽管变量名可以任意指定，但为提高可读性，最好使用有意义的单词。上面ca、cb都是不好的名称
- 必要时可以使用下划线，例如，chemist_name
- 可读性好的代码，便于调试，也方便他人使用

## 2.5 DataFrame，数据分析专用类型

上述数据类型都是在底层工作的，我们实际要处理的数据很想excel表格。
- 二维表，有行有列，行有编号1,2,3...，列有名称

- 需要整列运算数据
- 需要多行筛选、排序
- 需要把一个数据表拆分成多个
- 需要把多个数据表合并成一个

DataFrame应运而生，它包含了excel所有的数据处理功能，甚至更多。

### 2.3.3 DataFrame 排序
