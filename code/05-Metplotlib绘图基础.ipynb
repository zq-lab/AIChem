## 基础绘图 Metaplotlib
  为了便于演示，我们引入numpy用于产生数据。实际使用场合，数据来源于实验仪器或者网络上的数据库。numpy内置了很多与数学运算相关的内容。
  绘图包使用Matplotlib。它不仅适合入门，而且功能也很强劲，足以满足绝大数应用场合。

```python
import numpy as np
import matplotlib.pyplot as plt

# 产生-π到π的100个数据点 
x = np.linspace(-np.pi, np.pi,100) #产生了一个数列
yc,ys = np.cos(x), np.sin(x)

plt.plot(x,yc)
plt.plot(x,ys)
plt.show()
```

 Matplotlib中包含有默认设置，用于定义绘图细节和颜色配置方案，默认的绘图能满足基本需求，无需改变。Matplotlib绘图包含以下几个部分：
 - Figure
 - axes
 - axis（包含x轴、y轴，以及tick、tick label和label）
 - data
 - title
 - legend

```python
#升级绘图方案
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-np.pi, np.pi,100)
yc,ys = np.cos(x), np.sin(x)

plt.plot(x,yc)
plt.plot(x,ys)
plt.legend(['cos(x)','sin(x)'],loc='best') # 按上述plot顺序一一对应
plt.show()
```

```python
# 继续改进绘图效果
## 分辨率，点、线型等
## x、y轴细节

import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-np.pi, np.pi,100)
yc,ys = np.cos(x), np.sin(x)

plt.figure(dpi=600)
plt.xlabel('temperature (degC)')
plt.ylabel('weather index')

plt.plot(x,yc,marker='',linestyle='--',linewidth=1)
plt.plot(x,ys)

lgd = plt.legend(['cos(x)','sin(x)']) # 按上述plot顺序一一对应
frame = lgd.get_frame() 
frame.set_alpha(1) 
frame.set_facecolor('none')

plt.savefig('../img/tmp.png',format='png',transparent=True)
plt.show()
```

## 7.2 绘图类型

```python
import pandas as pd
df = pd.read_csv('data/zk_NP_coll.csv', encoding='utf-8')
df
```

## 7.2.1 条形图

```python
plt.figure(dpi = 120)
plt.xlabel('compound')
plt.ylabel('TPSA')
plt.bar(df.index,df['TPSA'])
plt.show()
```

### 7.2.2 直方图

```python
# 直方图
## 数据分布
plt.figure(dpi=300)
plt.xlabel('TPSA distribution')
plt.ylabel('Frequency')
plt.hist(df['TPSA'],bins=50)
plt.title('TPSA Distribution')
plt.show()
```

### 7.2.3 散点图

```python
# 散点图
plt.scatter(df.index,df['TPSA'],c='red')
plt.show()
```
